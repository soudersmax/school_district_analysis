```python
#Add Pandas dependency
import pandas as pd
import os
```

```python
#Files to Load
school_data_to_load = os.path.join("Resources", "schools_complete.csv")
student_data_to_load = os.path.join("Resources", "students_complete.csv")
```

```python
#Read the school data file and store it in a pandas DF
school_data_df = pd.read_csv(school_data_to_load)
school_data_df
```

```python
#Read the student data file and store it in a PandasDF
student_data_df = pd.read_csv(student_data_to_load)
student_data_df
```

```python
# Determine if there are any missing values in the school data.
school_data_df.count()
```

```python
# Determine if there are any missing values in the student data.
student_data_df.isnull().sum()
```

```python
#Determine data types for the school df
school_data_df.dtypes
```

```python
#Declare a list combining prefixes and suffixes of interest
prefix_suffix = ["Dr. ", "Mr. ","Mrs. ","Miss ","Ms. ", " MD"," DDS"," PhD", " DVM"]
```

```python
#iterate through the words in the prefix_suffix list and replace
for word in prefix_suffix:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
```

```python
student_data_df
```

```python
#Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df,school_data_df, on=["school_name","school_name"])
school_data_complete_df.head()
```

```python
#Get the total number of students
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
#Get the total number of schools
school_count = school_data_df["school_name"].count()
school_count
```

```python
#Calculate total budget
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
#Calculate average reading score
avg_reading = school_data_complete_df["reading_score"].mean()
avg_reading
```

```python
#Calculate average math score
avg_math = school_data_complete_df["math_score"].mean()
avg_math
```

```python
# Make lists of passing students
passing_math = school_data_complete_df["math_score"] >= 70
passing_reading = school_data_complete_df["reading_score"] >= 70
```

```python
# Filter data frame to get the students who are passing math in a new DataFrame.
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
```

```python
# Filter data frame to get the students who are passing reading in a new DataFrame.
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
```

```python
#calculate the number of students passing each subject
passing_math_count = passing_math["student_name"].count()
passing_reading_count = passing_reading["student_name"].count()
print(passing_math_count)
print(passing_reading_count)
```

```python
#calculate the percentage that passed each test
passing_math_percentage = passing_math_count / float(student_count) *100
passing_reading_percentage = passing_reading_count / float(student_count) *100
print(passing_math_percentage)
print(passing_reading_percentage)
```

```python
# Filter df by passing both math and reading
passing_both = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70) & (school_data_complete_df["math_score"] >= 70)]
passing_both.head()
```

```python
# Calculate the number of students in passing_both df
passing_both_count = passing_both["student_name"].count()
passing_both_count
```

```python
# Calculate overall passing percentage
overall_pass_percentage = passing_both_count / student_count *100
overall_pass_percentage
```

```python
#Create district_summary_df from dict where keys are column names and values are calculated metrics from above
district_summary_df = pd.DataFrame(
        [{"Total Schools": school_count,
         "Total Students": student_count,
         "Total Budget": total_budget,
         "Average Math Score": avg_math,
         "Average Reading Score": avg_reading,
         "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
         "% Overall Passing": overall_pass_percentage}])
district_summary_df
```

```python
# Format the "Total Students" to have the comma for a thousands separator.
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Students"]
```

```python
#Format the "Total Budget" to have the comma for thousands separator, a decimal with 2 places, and a dollar sign
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Total Budget"]
```

```python
#Format Remaining Columns
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["Average Math Score"] =  district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)

district_summary_df
```

```python
#School Summary Analysis
#Determine the school type
per_school_types = school_data_df.set_index(["school_name"])["type"]
#Determine the school size
per_school_counts = school_data_df.set_index(["school_name"])["size"]
#Determine total school budget
per_school_budget = school_data_df.set_index(["school_name"])["budget"]
#determine school budget per capita
per_school_capita = per_school_budget / per_school_counts
# Calculate the average test scores.
per_school_math = school_data_complete_df.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_complete_df.groupby(["school_name"]).mean()["reading_score"]
# Calculate the passing scores by creating a filtered DataFrame (see above ln 86)
per_school_passing_math = school_data_complete_df[(school_data_complete_df["math_score"] >=70)]
per_school_passing_reading = school_data_complete_df[(school_data_complete_df["reading_score"] >=70)]
#Get the number of students in each of the above filtered df
per_school_passing_math = per_school_passing_math.groupby(["school_name"]).count()["student_name"]
per_school_passing_reading = per_school_passing_reading.groupby(["school_name"]).count()["student_name"]
#Determine the Percentage of students passing each test
per_school_passing_math = per_school_passing_math / per_school_counts *100
per_school_passing_reading = per_school_passing_reading / per_school_counts *100
#Calculate overall passing percentage for each school
per_school_both = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70) & (school_data_complete_df["math_score"] >= 70)]
per_school_both = per_school_both.groupby(["school_name"]).count()["student_name"]
per_school_both_percent = per_school_both / per_school_counts *100
#add all the calculated data to new school_summary df
school_summary_df = pd.DataFrame({
    "School Type": per_school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": per_school_both_percent})
#clean  and format school_summary_df
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)
#Sort top 5 schools
top_schools = school_summary_df.sort_values(["% Overall Passing"], ascending=False)
#sort lowest 5 schools
bottom_schools = school_summary_df.sort_values(["% Overall Passing"], ascending=True)
```

```python
#Grade-level analysis by school
#get series for each grade level
ninth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "9th")]
tenth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "10th")]
eleventh_grade = school_data_complete_df[(school_data_complete_df["grade"] == "11th")]
twelfth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "12th")]
#get average math scores
ninth_grade_math_avg = ninth_grade.groupby(["school_name"]).mean()["math_score"]
tenth_grade_math_avg = tenth_grade.groupby(["school_name"]).mean()["math_score"]
eleventh_grade_math_avg = eleventh_grade.groupby(["school_name"]).mean()["math_score"]
twelfth_grade_math_avg = twelfth_grade.groupby(["school_name"]).mean()["math_score"]
#get average reading scores
ninth_grade_reading_avg = ninth_grade.groupby(["school_name"]).mean()["reading_score"]
tenth_grade_reading_avg = tenth_grade.groupby(["school_name"]).mean()["reading_score"]
eleventh_grade_reading_avg = eleventh_grade.groupby(["school_name"]).mean()["reading_score"]
twelfth_grade_reading_avg = twelfth_grade.groupby(["school_name"]).mean()["reading_score"]
#combine the series into df
math_scores_by_grade = pd.DataFrame({
               "9th": ninth_grade_math_avg,
               "10th": tenth_grade_math_avg,
               "11th": eleventh_grade_math_avg,
               "12th": twelfth_grade_math_avg})
reading_scores_by_grade = pd.DataFrame({
                "9th": ninth_grade_reading_avg,
                "10th": tenth_grade_reading_avg,
                "11th": eleventh_grade_reading_avg,
                "12th": twelfth_grade_reading_avg})
#format math df
math_scores_by_grade["9th"] = math_scores_by_grade["9th"].map("{:,.1f}".format)
math_scores_by_grade["10th"] = math_scores_by_grade["10th"].map("{:,.1f}".format)
math_scores_by_grade["11th"] = math_scores_by_grade["11th"].map("{:,.1f}".format)
math_scores_by_grade["12th"] = math_scores_by_grade["12th"].map("{:,.1f}".format)
#remove index name
math_scores_by_grade.index.name = None
#format reading df
reading_scores_by_grade["9th"] = reading_scores_by_grade["9th"].map("{:,.1f}".format)
reading_scores_by_grade["10th"] = reading_scores_by_grade["10th"].map("{:,.1f}".format)
reading_scores_by_grade["11th"] = reading_scores_by_grade["11th"].map("{:,.1f}".format)
reading_scores_by_grade["12th"] = reading_scores_by_grade["12th"].map("{:,.1f}".format)
#remove index name
reading_scores_by_grade.index.name = None
```

```python
#Per Capita Analysis
#Get descriptive statistics for per_school_capita
per_school_capita.describe()
#Create and name spending bins series
spending_bins = [0, 585, 630, 645, 675]
group_names = ["<$584", "$585-629", "$630-644", "$645-675"]
#cut and group the  df
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=group_names)
#Calculate averages for each column
spending_math_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
#Assemble and format DataFrame
spending_summary_df = pd.DataFrame({
    "Average Math Score":spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending})
spending_summary_df["Average Math Score"] = spending_summary_df["Average Math Score"].map("{:.1f}".format)
spending_summary_df["Average Reading Score"] = spending_summary_df["Average Reading Score"].map("{:.1f}".format)
spending_summary_df["% Passing Math"] = spending_summary_df["% Passing Math"].map("{:.0f}".format)
spending_summary_df["% Passing Reading"] = spending_summary_df["% Passing Reading"].map("{:.0f}".format)
spending_summary_df["% Overall Passing"] = spending_summary_df["% Overall Passing"].map("{:.0f}".format)
spending_summary_df
```

```python
#School Size analysis
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
#cut and group the  df
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], size_bins, labels=group_names)
# Calculate averages for the desired columns.
size_math_scores = school_summary_df.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = school_summary_df.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = school_summary_df.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = school_summary_df.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing = school_summary_df.groupby(["School Size"]).mean()["% Overall Passing"]
#assemble and format dataframe
size_summary_df = pd.DataFrame({
          "Average Math Score" : size_math_scores,
          "Average Reading Score": size_reading_scores,
          "% Passing Math": size_passing_math,
          "% Passing Reading": size_passing_reading,
          "% Overall Passing": size_overall_passing})
size_summary_df["Average Math Score"] = size_summary_df["Average Math Score"].map("{:.1f}".format)
size_summary_df["Average Reading Score"] = size_summary_df["Average Reading Score"].map("{:.1f}".format)
size_summary_df["% Passing Math"] = size_summary_df["% Passing Math"].map("{:.0f}".format)
size_summary_df["% Passing Reading"] = size_summary_df["% Passing Reading"].map("{:.0f}".format)
size_summary_df["% Overall Passing"] = size_summary_df["% Overall Passing"].map("{:.0f}".format)
size_summary_df
```
